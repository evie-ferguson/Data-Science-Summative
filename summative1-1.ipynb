5. Convert `Date` to a pandas timestamp. 
   > tip: use `pd.to_datetime()`

````python
# converting Data to pandas timestamp
df['Date'] = pd.to_datetime(df['Date'])
````

6. Filter the DataFrame on **one specific module** (your choice) and plot its **attendance rate over time**
   (x = Date, y = the module's average attendance).



````python
# filtering by Spanish module
history_df = df[df['Module Name'] == 'History']
# calculating attendance rate over time for history module
history_attendance = history_df.groupby(history_df['Date'].dt.date)['Attended'].mean()
````

````python
# plotting the attendance rate over time for history
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.plot(history_attendance, marker='o', linestyle='-')
plt.title('Average History Module Attendance Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Attendance Rate')
plt.grid()
plt.show()
````

7. **[For L6 students only]** Add a column, titled "Student Overall Attendance", with the average attendance for the student. Add a column, titled "Standardised Student Overall Attendance", with the z-scores of "Student Overall Attendance".

````python
# adding a column for "Student Overall Attendance Rate"
student_attendance_rate = df.groupby(df['Person Code'])['Attended'].transform('mean')
df['student overall attendance'] = student_attendance_rate
# and "Standardised Student Overall Attendance Rate"
from scipy.stats import zscore
df['standardised student overall attendance'] = zscore(df['student overall attendance'])
df.head()
````

### **Exercise 4 â€” Comparing Modules**



1. Compute the **mean** and **standard deviation** of attendance per module using pandas `groupby`. Consult the documentation if needed.


````python
# computing mean and standard deviation of attendance per module
module_mean = df.groupby('Module Name')['Attended'].mean().reset_index(name='Mean Attended')
module_mean.head()
````

````python
module_stats = df.groupby('Module Name')['Attended'].agg(['mean', 'std']).reset_index().rename(columns={'mean': 'Mean Attended', 'std': 'Standard Deviation'})
module_stats
````

2. Create a `seaborn.barplot` of mean attendance per module with 95% confidence intervals.


````python
import seaborn as sns
sns.set_theme(style="whitegrid", rc={'figure.figsize':(24,8)})
sns.barplot(x="Module Name", y="Attended", data=df, estimator=np.mean, ci=95, capsize=0.05)
plt.xticks(rotation=90)
plt.show()
````

### **Exercise 5 â€” Attendance Relationships**



1. Answer the question: **Do students with low overall attendance tend to attend modules that have low attendance?**.


2. **[For L5 students only]**. Choose between a correlation analysis or a linear regression analysis to answer the question.

3. **[For L6 students only]** Run both correlation analysis and linear regression analysis to answer the same question. Do they tell the same story?

> Tips:
> 1. Compute the mean attendance rate of each module. Can you re-use some of the `groupby` code you wrote above?
> 2. For each module, compute the average overall attendance of the students who attended that module.
> 3. Correlation and linear regression can both help you answer the question. 
> 4. Report main statistic and p-value(s).


````python
# each persons average attendance for each module
# x - module attendance, y - overall attendance
````

````python
# Do students with low overall attendance tend to attend modules that have low attendance?
# sort modules based on lowest-highest attended
module_stats = module_stats.sort_values(by='Mean Attended',ascending=True)
module_stats
# sort student overall attendance from lowst to highest
# find correlation
````

````python
# creating df from dropped_duplicates with student overall attendance 
student_module = df[['Person Code', 'Module Name', 'student overall attendance']].drop_duplicates(subset=['Person Code', 'Module Name'])
# finding the average students overall attendance in that module
module_student_overall = student_module.groupby('Module Name')['student overall attendance'].mean().reset_index(name='Mean Student Overall Attendance')
# merging with module_stats so in one df
merged = module_stats.merge(module_student_overall, on='Module Name', how='inner')
merged.head(20)
````

````python
sns.regplot(data=merged, x= 'Mean Attended', y='Mean Student Overall Attendance', line_kws={'color':'orange'})
for _, row in merged.iterrows():
    plt.annotate(row['Module Name'],
                 xy=(row['Mean Attended'], row['Mean Student Overall Attendance']),
                 xytext=(5,5), textcoords='offset points',
                 fontsize=8)
plt.title("Relationship between mean attendance per module and student overall attendance")
plt.show()
````
